```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('INR=X.csv')
x = data[["Open", "High", "Low"]]
y = data["Close"]
x = x.to_numpy()
y = y.to_numpy()
y = y.reshape(-1, 1)

# Split data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

# Decision Tree model
dt_model = DecisionTreeRegressor()
dt_model.fit(xtrain, ytrain)
dt_ypred = dt_model.predict(xtest)
dt_mse = np.mean((dt_ypred - ytest)**2)
print('Decision Tree Test MSE:', dt_mse)

# Artificial Neural Network model
ann_model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(3,)),
    keras.layers.Dense(1)
])
ann_model.compile(optimizer='adam', loss='mse')
ann_model.fit(xtrain, ytrain, epochs=100, batch_size=32, verbose=0)
ann_ypred = ann_model.predict(xtest)
ann_mse = np.mean((ann_ypred - ytest)**2)
print('Artificial Neural Network Test MSE:', ann_mse)

# Linear Regression model
lr_model = LinearRegression()
lr_model.fit(xtrain, ytrain)
lr_ypred = lr_model.predict(xtest)
lr_mse = np.mean((lr_ypred - ytest)**2)
print('Linear Regression Test MSE:', lr_mse)

# Plot predictions against actual exchange rates for all models
plt.scatter(ytest, dt_ypred, label='Decision Tree')
plt.scatter(ytest, ann_ypred, label='Artificial Neural Network')
plt.scatter(ytest, lr_ypred, label='Linear Regression')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)
plt.xlabel('Actual Exchange Rate')
plt.ylabel('Predicted Exchange Rate')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('INR=X_new.csv')
x = data[["Open", "High", "Low"]]
y = data["Close"]
x = x.to_numpy()
y = y.to_numpy()
y = y.reshape(-1, 1)

# Split data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

# Decision Tree model
dt_model = DecisionTreeRegressor()
dt_model.fit(xtrain, ytrain)
dt_ypred = dt_model.predict(xtest)
dt_mse = np.mean((dt_ypred - ytest)**2)
print('Decision Tree Test MSE:', dt_mse)

# Artificial Neural Network model
ann_model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(3,)),
    keras.layers.Dense(1)
])
ann_model.compile(optimizer='adam', loss='mse')
ann_model.fit(xtrain, ytrain, epochs=100, batch_size=32, verbose=0)
ann_ypred = ann_model.predict(xtest)
ann_mse = np.mean((ann_ypred - ytest)**2)
print('Artificial Neural Network Test MSE:', ann_mse)

# Linear Regression model
lr_model = LinearRegression()
lr_model.fit(xtrain, ytrain)
lr_ypred = lr_model.predict(xtest)
lr_mse = np.mean((lr_ypred - ytest)**2)
print('Linear Regression Test MSE:', lr_mse)

# Plot predictions against actual exchange rates for all models
sns.set_style('whitegrid')
fig, axs = plt.subplots(ncols=3, figsize=(15,5))
axs[0].scatter(ytest, dt_ypred, alpha=0.5)
sns.regplot(x=ytest, y=dt_ypred, ax=axs[0], color='g')
axs[0].set_title('Decision Tree')
axs[0].set_xlabel('Actual Exchange Rate')
axs[0].set_ylabel('Predicted Exchange Rate')
axs[1].scatter(ytest, ann_ypred, alpha=0.5)
sns.regplot(x=ytest, y=ann_ypred, ax=axs[1], color='b')
axs[1].set_title('Artificial Neural Network')
axs[1].set_xlabel('Actual Exchange Rate')
axs[1].set_ylabel('Predicted Exchange Rate')
axs[2].scatter(ytest, lr_ypred, alpha=0.5)
sns.regplot(x=ytest, y=lr_ypred, ax=axs[2], color='r')
axs[2].set_title('Linear Regression')
axs[2].set_xlabel('Actual Exchange Rate')
axs[2].set_ylabel('Predicted Exchange Rate')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('INR=X_new.csv')
x = data[["Open", "High", "Low"]]
y = data["Close"]
x = x.to_numpy()
y = y.to_numpy()
y = y.reshape(-1, 1)

# Split data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

# Decision Tree Regression
dt_model = DecisionTreeRegressor()
dt_model.fit(xtrain, ytrain)
dt_ypred = dt_model.predict(xtest)
dt_mse = np.mean((dt_ypred - ytest) ** 2)
print('Decision Tree Test MSE:', dt_mse)

# Artificial Neural Network
scaler = StandardScaler()
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)

ann_model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(3,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(1)
])

ann_model.compile(optimizer='adam', loss='mse')
ann_model.fit(xtrain_scaled, ytrain, epochs=100, batch_size=32, verbose=0)
ann_ypred = ann_model.predict(xtest_scaled)
ann_mse = np.mean((ann_ypred - ytest) ** 2)
print('Artificial Neural Network Test MSE:', ann_mse)

# Linear Regression
lr_model = LinearRegression()
lr_model.fit(xtrain, ytrain)
lr_ypred = lr_model.predict(xtest)
lr_mse = np.mean((lr_ypred - ytest) ** 2)
print('Linear Regression Test MSE:', lr_mse)

# Print the first 10 predicted values for each model
print('Decision Tree Predictions:')
for prediction in dt_ypred[:10]:
    print(prediction)

print('Artificial Neural Network Predictions:')
for prediction in ann_ypred[:10]:
    print(prediction[0])

print('Linear Regression Predictions:')
for prediction in lr_ypred[:10]:
    print(prediction)

# Plot predictions against actual exchange rates for all models
plt.scatter(ytest, dt_ypred, label='Decision Tree')
plt.scatter(ytest, ann_ypred, label='Artificial Neural Network')
plt.scatter(ytest, lr_ypred, label='Linear Regression')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)
plt.xlabel('Actual Exchange Rate')
plt.ylabel('Predicted Exchange Rate')
plt.legend()
plt.show()
```
